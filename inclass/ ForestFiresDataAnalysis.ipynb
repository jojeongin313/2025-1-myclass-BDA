## 산불건수와 강수량의 상관관계 (비교)

```python
import matplotlib.pyplot as plt

def plot_fire_rain_by_region(region_name):
    df_region = df[df['시도'] == region_name]

    if df_region.empty:
        print(f"'{region_name}'에 해당하는 데이터가 없습니다.")
        return

    fig, ax1 = plt.subplots(figsize=(10, 5))

    color1 = 'tab:red'
    ax1.set_xlabel('연도')
    ax1.set_ylabel('산불건수', color=color1)
    ax1.plot(df_region['연도'], df_region['산불건수'], color=color1, label='산불건수')
    ax1.tick_params(axis='y', labelcolor=color1)

    ax2 = ax1.twinx()
    color2 = 'tab:blue'
    ax2.set_ylabel('강수량', color=color2)
    ax2.plot(df_region['연도'], df_region['강수량'], color=color2, label='강수량')
    ax2.tick_params(axis='y', labelcolor=color2)

    plt.title(f'{region_name} - 연도별 산불건수 vs 강수량')
    fig.tight_layout()
    plt.show()


region = input("지역명을 입력하세요: ")
plot_fire_rain_by_region(region)
```

강수량이 극히 낮은 년도에는 **산불**이 많이 났음

## 강수량과 산불건수의 상관관계 (파악)

```python
# 시도별 평균 계산
avg_by_region = df.groupby('시도')[['강수량', '산불건수']].mean().rename(columns={'강수량': '평균강수량', '산불건수': '평균산불건수'})

# 원본 데이터에 병합
dfalpha = df.merge(avg_by_region, on='시도')
```

```python
dfalpha['강수량많음'] = dfalpha['강수량'] > dfalpha['평균강수량']
dfalpha['강수량적음'] = dfalpha['강수량'] < dfalpha['평균강수량']
dfalpha['산불많음'] = dfalpha['산불건수'] > dfalpha['평균산불건수']
dfalpha['산불적음'] = dfalpha['산불건수'] < dfalpha['평균산불건수']
```

```python
# 조건별 건수 계산
data = [
    ['강수량많고', '산불많음', ((dfalpha['강수량많음']) & (dfalpha['산불많음'])).sum()],
    ['강수량많고', '산불적음', ((dfalpha['강수량많음']) & (dfalpha['산불적음'])).sum()],
    ['강수량적고', '산불많음', ((dfalpha['강수량적음']) & (dfalpha['산불많음'])).sum()],
    ['강수량적고', '산불적음', ((dfalpha['강수량적음']) & (dfalpha['산불적음'])).sum()],
]

condition_counts = pd.DataFrame(data, columns=['강수량', '산불', '건수'])

# 피벗 테이블
pivot_table = condition_counts.pivot_table(index='강수량', columns='산불', values='건수', aggfunc='sum')

# 시각화
plt.figure(figsize=(6, 4))
sns.heatmap(pivot_table, annot=True, cmap="YlOrRd", fmt=".0f")
plt.title("강수량-산불 조건별 데이터 건수 Heatmap")
plt.xlabel("산불 조건")
plt.ylabel("강수량 조건")
plt.show()
```

단순히 강수량이 많다고 해서 산불이 억제된다고 보긴 어려움
(계절적 요인, 바람, 지형 등 다른 변수 영향 가능성)

**강수량이 적을수록 산불 위험이 높은 경향**

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 시도별 강수량, 산불건수 합계
summary = dfalpha.groupby('시도')[['산불건수', '강수량']].sum()

# 2. 평균 산불건수 계산 및 산불많음 여부
avg_fire = summary['산불건수'].mean()
summary['산불많음'] = summary['산불건수'] > avg_fire

# 3. 데이터 분리
fire_high = summary[summary['산불많음'] == True].sort_values(by='강수량', ascending=False)
fire_low = summary[summary['산불많음'] == False].sort_values(by='강수량', ascending=False)

# 4. 시각화
plt.figure(figsize=(14, 6))

# 산불 많은 지역 (빨간색)
plt.bar(fire_high.index, fire_high['강수량'], color='orangered', label='산불 많음')

# 산불 적은 지역 (파란색)
plt.bar(fire_low.index, fire_low['강수량'], color='skyblue', label='산불 적음')

# 5. 설정
plt.title('시도별 강수량 총합 (산불 많음 여부 색상)', fontsize=14)
plt.xlabel('시도')
plt.ylabel('강수량 총합 (mm)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.legend()

plt.tight_layout()
plt.show()
```

경기도와 경상남도, 강원특별자치도 -> 강수량&산불건수 모두 많음<br>

## 강수량이 많을수록 산불면적이 작아지는 경향이 뚜렷한 지역

```python
# 1. 데이터 불러오기
df2 = pd.read_excel('산불통계자료.xlsx')
df2.columns = ['지역', '연도', '산불건수', '산불면적', '강수량']
```

```python
# 2. 결측값 전처리
df2['강수량'] = pd.to_numeric(df2['강수량'].replace('-', pd.NA), errors='coerce')
df2['산불면적'] = pd.to_numeric(df2['산불면적'].replace('-', pd.NA), errors='coerce')
```

```python
# 3. 지역별 음의 상관관계 계산
corr_list = []
for 지역 in df2['지역'].unique():
    sub_df = df2[df2['지역'] == 지역]
    if len(sub_df) >= 3:
        corr = sub_df['강수량'].corr(sub_df['산불면적'])
        corr_list.append({'지역': 지역, '음의 상관계수': corr})

corr_df = pd.DataFrame(corr_list)
neg_corr_df = corr_df.sort_values('음의 상관계수')
```

```python
# 4. 음의 상관관계가 큰 지역 Top 3
top3_neg_corr_regions = neg_corr_df.head(3)['지역'].tolist()
```

```python
# 5. 시각화
for 지역 in top3_neg_corr_regions:
    region_df = df2[df2['지역'] == 지역]

# 산점도
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=region_df, x='강수량', y='산불면적')
    plt.title(f"[{지역}] 강수량과 산불면적 간 음의 상관관계")
    plt.xlabel("강수량 (mm)")
    plt.ylabel("산불면적 (ha)")
    plt.tight_layout()
    plt.show()
```

```python
for 지역 in top3_neg_corr_regions:
    region_df = df2[df2['지역'] == 지역]

# 선 그래프
    plt.figure(figsize=(8, 4))
    sns.lineplot(data=region_df, x='연도', y='강수량', label='강수량', marker='o')
    sns.lineplot(data=region_df, x='연도', y='산불면적', label='산불면적', marker='s')
    plt.title(f"[{지역}] 연도별 강수량 및 산불면적 추이")
    plt.ylabel("값")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
